```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import poisson,skellam
import os
import sys
import ipywidgets as widgets
%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8) # taille des figures

package_dir = os.path.dirname(os.path.abspath(sys.argv[1]))
file_name = "FullChampionnat"
thefile = os.path.join(package_dir, file_name+'.csv')
ext = pd.read_csv(thefile)
```

```python
# Importing re package for using regular expressions
import re
  
# Function to clean the names
def Clean_names(City_name):
    # Search for opening bracket in the name followed by
    # any characters repeated any number of times
    if re.search('\(.*', City_name):
  
        # Extract the position of beginning of pattern
        pos = re.search('\(.*', City_name).start()
  
        # return the cleaned name
        return City_name[:pos]
  
    else:
        # if clean up needed return the same name
        return City_name
          
# Updated the city columns
ext['team_a'] = ext['team_a'].apply(Clean_names)
ext['team_b'] = ext['team_b'].apply(Clean_names)
ext['team_b'] = ext['team_b'].str.replace(' ', '')
ext['team_a'] = ext['team_a'].str.replace(' ', '')
ext['team_b'] = ext['team_b'].str.replace('\xa0', '')
ext['team_a'] = ext['team_a'].str.replace('\xa0', '')
Ligue = list(ext['team_b'].unique()) + list(ext['team_a'].unique())
Ligue = set(Ligue)
Championnat = list(ext['Championnat'].unique())
Championnat = set(Championnat)
clean_ligue = list(set(Ligue))
cleanequipe = clean_ligue
cleanequipe.sort()
cleanequipe
```

## <center>Resultat Match à Domicile </center>

```python
def say_my_name(Domicile,Exterieur,Championnat):
    """
    """
    try:
        ext.loc[ext['score_a'] > ext['score_b'], 'resul'] = '1' 
        ext.loc[ext['score_a'] == ext['score_b'], 'resul'] = '0' 
        ext.loc[ext['score_a'] < ext['score_b'], 'resul'] = '2' 
        df3 = ext[ext['team_a'] == Domicile]
        df3 = df3[df3['team_b'] == Exterieur]
        df3 = df3[df3['Championnat'] == Championnat]
        
        df3 = df3.reset_index()
        Match = len(df3)
        Win = df3[df3.resul == '1']
        Loose = df3[df3.resul == '2']
        Null = df3[df3.resul == '0']
        P_Win = len(Win)/Match*100
        P_Null= len(Null)/Match*100
        P_Loose = len(Loose)/Match*100
        #define data
        data = [round(P_Win,2), round(P_Null,2), round(P_Loose,2)]
        labels = ['Victoire', 'NULL', 'Loose']

        #define Seaborn color palette to use
        colors = sns.color_palette('pastel')[0:5]

        #create pie chart
        plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
        plt.show()
        print(f"Myenne de but équipe Domicile : {round(df3['score_a'].mean(),2)}")
        print(f"Moyenne de but équipe Exterieur : {round(df3['score_b'].mean(),2)}")

    except :
        print("Veuillez indiquer deux équipes différente")



widgets.interact(say_my_name, Domicile=cleanequipe,Exterieur = cleanequipe,Championnat=Championnat);
```
